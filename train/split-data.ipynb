```python
import os
import shutil
import random

# Direktori di mana 9 folder baru disimpan
main_dir = '/path/to/output_dataset'
# Direktori output untuk train dan test set
train_dir = '/path/to/train_dataset'
test_dir = '/path/to/test_dataset'
# Persentase untuk data training dan testing
train_split = 0.8

# Buat direktori train dan test jika belum ada
if not os.path.exists(train_dir):
    os.makedirs(train_dir)
if not os.path.exists(test_dir):
    os.makedirs(test_dir)

# Dapatkan daftar 9 folder baru
folders = os.listdir(main_dir)

# Proses setiap folder
for folder in folders:
    folder_path = os.path.join(main_dir, folder)
    images = os.listdir(folder_path)
    
    # Acak urutan gambar
    random.shuffle(images)
    
    # Tentukan jumlah data training
    train_size = int(len(images) * train_split)
    
    # Ambil data untuk training dan testing
    train_images = images[:train_size]
    test_images = images[train_size:]
    
    # Buat folder baru untuk data training dan testing sesuai dengan folder asalnya
    train_folder = os.path.join(train_dir, folder)
    test_folder = os.path.join(test_dir, folder)
    
    if not os.path.exists(train_folder):
        os.makedirs(train_folder)
    if not os.path.exists(test_folder):
        os.makedirs(test_folder)
    
    # Salin gambar ke direktori train
    for image in train_images:
        src = os.path.join(folder_path, image)
        dst = os.path.join(train_folder, image)
        shutil.copyfile(src, dst)
    
    # Salin gambar ke direktori test
    for image in test_images:
        src = os.path.join(folder_path, image)
        dst = os.path.join(test_folder, image)
        shutil.copyfile(src, dst)

print("Proses split data selesai!")
```
